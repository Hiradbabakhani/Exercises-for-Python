```python
class Account:
    def __init__(self, owner, balance=0):
        self.owner = owner
        self.__balance = balance

    def deposit(self, amount):
        self.__balance += amount
        print(f'Added {amount} to the balance')

    def show_balance(self):
        print(f'Balance {self.__balance}')

    def withdraw(self, amount):
        if amount > self.__balance:
            print('Not enough funds')
        else:
            self.__balance -= amount
            print(f'withdraw {amount} from the balance')
            
```

```python
hirad_acc = Account('hirad',2600)
```

```python
hirad_acc.show_balance()
```

```python
class Dog():
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def bark(self):
        return f'{self.name} says woof!'
    
    def run(self):
        return f'{self.name} can run at this {self.age} age'
    
    def sleep(self):
        return f'{self.name} slepping! pleas dont speac.'
```

```python
dog1 = Dog('Max',1)
dog2 = Dog('Jacki',1.5)
dog3 = Dog('Rocki',3)
```

```python
dog1.bark()
```

```python
dog2.run()
```

```python
dog3.sleep()
```
