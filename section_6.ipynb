```python
class Account:
    def __init__(self, owner, balance=0):
        self.owner = owner
        self.__balance = balance

    def deposit(self, amount):
        self.__balance += amount
        print(f'Added {amount} to the balance')

    def show_balance(self):
        print(f'Balance {self.__balance}')

    def withdraw(self, amount):
        if amount > self.__balance:
            print('Not enough funds')
        else:
            self.__balance -= amount
            print(f'withdraw {amount} from the balance')
            
```

```python
hirad_acc = Account('hirad',2600)
```

```python
hirad_acc.show_balance()
```

```python
class Dog():
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def bark(self):
        return f'{self.name} says woof!'
    
    def run(self):
        return f'{self.name} can run at this {self.age} age'
    
    def sleep(self):
        return f'{self.name} slepping! pleas dont speac.'
```

```python
dog1 = Dog('Max',1)
dog2 = Dog('Jacki',1.5)
dog3 = Dog('Rocki',3)
```

```python
dog1.bark()
```

```python
dog2.run()
```

```python
dog3.sleep()
```

```python
class Book:
    book_count = 0 
    

    def __init__(self, title, author, pages):
        self.title = title
        self.author = author
        self.pages = pages
    
    # We need to use this method to ensure that the output we get is correct, and we use this word to reference accurately.
    def description(self):
        print(f'title: {self.title}')
        print(f'author: {self.author}')
        print(f'pages: {self.pages}')
        Book.book_count += 1

    def is_long(self):
        return self.pages > 500
    

book1 = Book('Python Programming', 'john Doe', 350)
book2 = Book('Django for Beginners', 'Williams vincent',308 )
book3 = Book('Clean code', 'Robert martin', 620)

book1.description()

book2.description()

book3.description()

print(f'total book: {Book.book_count}')

print(f'This book is considered long {book3.is_long()}')
```
